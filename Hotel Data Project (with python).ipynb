## <span style= 'color: red'> Import Libraries And Data</span>

```python
# importing libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)



# loading data

df2018 = pd.read_csv(r'C:\Users\DELL LATITUDE E5470\Documents\Dataframe\Hotel Data\2018.csv')

df2019 = pd.read_csv(r'C:\Users\DELL LATITUDE E5470\Documents\Dataframe\Hotel Data\2019.csv')

df2020 = pd.read_csv(r'C:\Users\DELL LATITUDE E5470\Documents\Dataframe\Hotel Data\2020.csv')


meal = pd.read_csv(r'C:\Users\DELL LATITUDE E5470\Documents\Dataframe\Hotel Data\meal.csv')


market = pd.read_csv(r'C:\Users\DELL LATITUDE E5470\Documents\Dataframe\Hotel Data\market.csv')
```

## <span style= 'color: red'> Data Cleaning And Manipulation</span>

In this phase, I will be performing the following operation:

- Inspect the whole data.
- Merge all the dataframes together for easily and accurate analysis.
- Convert columns to it's right data type to make analysis easy.
- Organize data to make it easier to read.

```python
### setting instructions to enable the full number of columns of the dataframe to be previewed

pd.set_option('display.max_columns', None)



### joining 2018, 2019 and 2020 tables together

hotels= pd.concat([df2018,df2019,df2020], ignore_index= True)

hotels
```

```python
### merging hotel table with meal and market segment tables

hotels = pd.merge(hotels, market)

hotels = pd.merge(hotels, meal)


hotels
```

**Columns for Revenue and Total Nights are needed for proper analysis.**

```python
### checking data types

hotels.dtypes
```

**"reservation_status_date" column needs to be changed to a datetime data type.**

```python
### checking unique names for the hotel column

len(hotels['hotel'].unique())
```

**2 unique names for the "hotel" column were counted.**

```python
### converting "reservation_status_date" to a datetime64 data type

hotels['reservation_status_date'] = pd.to_datetime(hotels['reservation_status_date'], format='%m/%d/%Y')



### confirming changes

hotels.dtypes
```

**"reservation_status_date" has been converted to a datetime data type.**

```python
### creating a new column for revenue

revenue = ((hotels['stays_in_week_nights'] + hotels['stays_in_weekend_nights']) * hotels['adr'])* hotels['Discount']

hotels['revenue'] = revenue




### creating a new column for total_nights

total_nights= (hotels['stays_in_week_nights'] + hotels['stays_in_weekend_nights'])
               
hotels['total_nights']= total_nights


hotels
```

**2 new columns (revenue, total_nights) have been created.**  

## <span style= 'color: red'> ANALYSIS & VISUALIZATION </span>

```python
### identifying the rows needed for summary

rev= hotels.revenue

adr=hotels.adr

hot=hotels.hotel

tot=hotels.total_nights

park=hotels.required_car_parking_spaces

dis=hotels.Discount



### sum of certain columns

rev= rev.sum(axis=0)

tot= tot.sum(axis=0)

park= park.sum(axis=0)



### average of certain columns

adr= adr.mean()

dis= (dis.mean())*100
```

```python
### creating a dataframe called summary

summary= {'Revenue' : rev, 'Avg. Adr': adr, 'Avg. Discount' : dis, 'Required Car Parking Spaces': park, 'Total Nights' : tot}

summary= pd.DataFrame(summary, index=['>'])

summary= summary.round(1)
```

```python
### changing all the datatypes to string so currency and percentage can be added

summary= summary.astype(str)

summary.dtypes
```

```python
### editting how the table should appear

summary['Revenue'] = '$' +summary['Revenue']

summary['Avg. Adr'] ='$'+summary['Avg Adr']

summary['Total Nights'] = summary['Total Nights']

summary['Required Car Parking Spaces'] =summary['Required Car Parking Spaces']

summary['Avg. Discount'] = summary['Avg Discount'] +'%'

summary
```

**New dataframe shows a summary of the data we need.** 

```python
### creating the range of date values we need

start_date = pd.to_datetime('2018-01-01')


end_date= pd.to_datetime('2020-12-31')

hotelss= hotels.loc[(hotels['reservation_status_date']> start_date) & (hotels['reservation_status_date'] < end_date)]

hotelss
```

**Reservation date has successfully been adjusted to a date range of 2018 to 2020 because the date before 2018 isn't consistent.**  

```python
### plotting line graph

sns.set_style('darkgrid')

sns.lineplot(x='reservation_status_date', y='revenue', data= hotelss, hue= 'hotel',  palette= ['red','yellow'], linewidth= 2.5)

plt.title('Revenue By Reservation Status Date')

plt.xlabel('Reservation Status Date (2018 - 2020)')

plt.ylabel('Revenue')

plt.show()
```

##### Key Findings

From the above line graph, I discovered that:
- Resort Hotel made the most revenue in 2018.

- Resort Hotel made higher revenue in single days.

```python
### trying to get the sum of city hotel's revenue

city= hotelss[hotelss['hotel'] == 'City Hotel']

sum_city= city['revenue'].sum()



### trying to get the sum of resort hotel's revenue

resort= hotelss[hotelss['hotel'] == 'Resort Hotel']

sum_resort= resort['revenue'].sum()
```

```python
# plotting pie chart

slices = [sum_city, sum_resort]

labels = ['City Hotel', 'Resort Hotel']

colours = ['yellow', 'red']

explode = [0, 0.1]

plt.style.use('default')

plt.pie(slices, labels = labels, 
        colors = colours, wedgeprops = {'edgecolor': 'Black'}, 
        explode = explode, autopct = '%1.1f%%')

plt.title('Piechart For Revenue Of Each Hotel')

plt.tight_layout()

plt.show()
```

##### Key Findings

From the above pie chart, I discovered that:
- City Hotel made the most revenue in total.
