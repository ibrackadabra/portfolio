## <span style= 'color: brown'> Data Cleaning And Manipulation</span>

In this phase, I will be performing the following operation:
- Remove null and incorrect values.
- Convert certain columns to it's right data type to make analysis easy.
- Create new columns.
- Organize data to make it easier to read.

```python
# removing missing data

movies= df.dropna(how = 'any')
```

```python
# inspecting missing data again 

movies.isnull().sum()
```

**No missing value detected from inspection.**

```python
# changing data type in  certain columns

movies= movies.astype({'budget' : 'int64', 'gross' : 'int64', 'votes' : 'int64', 'runtime' : 'int64' })
movies.info()
```

**Budget, Gross, Votes and Runtime columns have been converted to the right data type.**

```python
# splitting "released" column into "date" and "country" columns

movies[['date', 'country']] = movies['released'].str.split('(', expand=True)



# removing the closing bracket from country

movies['country'] = movies['country'].astype(str).str.replace(')', '')
movies.head(n=100)
```

**New columns have been created for easy analysis.**

```python
# removing potential leading/trailing whitespaces from Date & Country

stripCol = ['date', 'country']

for col in stripCol:
    movies[col] = movies[col].astype(str).str.strip()
```

```python
# deleting "released" column because it's not needed again

del movies['released']
```

```python
# deleting "year" column since we already have "date" column 

del movies['year']
movies.head(n=100)
```

**Released and Year columns have successfully been deleted.**

```python
# checking certain rows because they contain incorrect data

movies.loc[[312, 800, 1173, 2816, 4187],:]
```

**I noticed some rows did not have a month in the Date column.**

```python
# deleting certain rows because they contain incorrect data

movies.drop([312, 800, 1173, 2816, 4187], axis=0, inplace=True)
```

**Rows without month in the Date column have successfully been deleted.**

```python
# inspecting data for duplicates

movies['company'].duplicated().sum()
```

**3944 duplicates found.**

```python
# dropping duplicates in company column

movies['company'].drop_duplicates()
```

**Duplicates successfully removed for easy analysis.**

```python
# sorting data by gross

movies = movies.sort_values('gross', ascending=False)
movies
```

```python
# converting date column to a datetime data type

movies['date'] =  pd.to_datetime(movies.date)
movies.info()
```

```python
# creating column for the month movies were released

movies['month']= movies['date'].dt.month_name()

movies
```

**Month column successfully created.**

```python
# sorting data by months

sort = ['January', 'February', 'March', 'April', 'May', 'June','July','August', 'September', 'October', 'November', 'December']


movies.index= pd.CategoricalIndex(movies['month'], categories= sort, ordered= True)
```

**Dataframe successfully sorted from highest to lowest gress values.**

## <span style= 'color: brown'> ANALYSIS & VISUALIZATION </span>

```python
# building scatter plot with budget vs gross

plt.scatter(x=movies['budget']/10**6, y=movies['gross']/10**6)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
# plotting budget vs gross using seaborn

sns.regplot(x= 'budget', y='gross', data=movies, scatter_kws={'color': 'red'}, line_kws={'color': 'black'})
```

**Movies with high budget have high gross values.**

```python
# checking correlation (numerical data only)

corr= movies.corr()
corr
```

**Numeric proof that movies with high budget have high gross earnings.** 

```python
# plotting Heatmap of our correlated data (numerical data only)

sns.heatmap(corr, annot=True)
plt.title('Correlation Between Data')
```

**This shows movies with high budget and high votes have high gross values.**

```python
# checking for correlation between data (with non-numerical data)



# first we will convert the nonnumerical data to category

dfn= df

for col_name in dfn.columns:
    if(dfn[col_name].dtype == 'object'):
        dfn[col_name] = dfn[col_name].astype('category')
        dfn[col_name] = dfn[col_name].cat.codes
        
dfn
```

```python
# plotting Heatmap for our correlated data

corrn= dfn.corr()
corrn
sns.heatmap(corrn, annot=True)
plt.title('Correlation Between Data (non-numerical data)')
```

**Other columns beside budget and vote have lower correlation to gross.**

```python
# showing all the high correlating columns to gross

corrn_pairs= corrn.unstack()
corrn_pairs
```

```python
sorted_pairs = corrn_pairs.sort_values()
sorted_pairs.head(n=50)
```

```python
high_corr= sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```

**further proves that Other columns beside budget and vote have lower correlation to gross.**
